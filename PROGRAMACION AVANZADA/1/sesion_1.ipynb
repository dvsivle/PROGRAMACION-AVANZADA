```python
import requests  #solicita a una pagina web que responda una consulta

source = 'https://pomber.github.io/covid19/timeseries.json'

r = requests.get(source)   #devuelve la respuesta a la consulta : 404 - Error, 200 - todo está bien (el archivo existe)
print(r)
print(type(r))
```

```python
json_data = r.json()  # retorna un diccionario simepre y cuando r haya sido cargado

type(json_data)
```

```python
print(json_data['Peru'][-1])
```

```python
#¿Cuántos países existen en el archivo?
country_list = [country for country in json_data.keys()]
country_list.sort()   # no devuelve una lista, solo ordena la lista a la que se aplica
```

```python
for idx, country in enumerate(country_list, start=1):
    print(f'{idx:3}: {country}')
    
print(f'Número de paises: {len(country_list)}')
```

```python
data_peru = json_data['Peru'][-1]
print(data_peru)

data_peru = json_data['peru'][-1]
print(data_peru)
```

```python
data_peru = json_data.get('Peru', 'Este país no existe')
print(data_peru)
```

```python
for data in data_peru:
    for k, v in data.items():   # items() devuelve tuplas
        print(f'{k.upper():5}: {v} \t', end = '')
    else:
        print()
```
