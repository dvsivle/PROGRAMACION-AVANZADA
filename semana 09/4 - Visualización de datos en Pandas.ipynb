# Visualización de datos en Pandas
La combinación de las librerías Pandas y Matplotlib permiten obtener gráficos a partir de los DataFrames. Aunque es posible utilzar los datos de un DataFrame como argumentos de las instrucciones para construír gráficos de Matplotlib, Pandas accede al Matplotlib API (esto es, a utilizar directamente las funciones gráficas de Matplotlib) de forma tal que puede utilizar sus propios métodos para generar los graficos resultantes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')      # Modificamos el formato del estilo para obtener gráficos más atractivos. Default:('classic')
```

## Pandas plot
El método df.plot() permite gráficar un DataFrame en una figura con varias opciones:

```python
df = pd.DataFrame(data=np.random.randn(365), #365 valores entre -1 y 1
                 index=pd.date_range('2019/1/1', periods=365), #crea un rango de fechas
                 columns=['Stock Price'])
df = df.cumsum() #suma acumulada
df.plot(figsize=(8, 6))
plt.show()
```

```python
df = pd.DataFrame(data=np.random.randn(365, 4), #365 valores en cada columna(4)
                 index=pd.date_range('2019/1/1', periods=365), 
                 columns=['Stock 1', 'Stock 2', 'Stock 3', 'Stock 4'])
df = df.cumsum()
df.plot(figsize=(8, 6))
plt.show()
```

```python
df = pd.DataFrame(data=np.random.randn(365, 2), 
                 columns=['Stock 1', 'Stock 2']).cumsum()
df['index'] = pd.Series(data=list(range(len(df))))
df.head() # muestra lo obtenido

df.plot(x='index', y='Stock 1', color='g', figsize=(8, 6))
plt.show()
```

```python
df = pd.DataFrame(data=np.random.randn(365, 2), 
                 columns=['Stock 1', 'Stock 2']).cumsum()
df['index'] = pd.Series(data=list(range(len(df))))
df.head()

df.plot(x='index', y=['Stock 1', 'Stock 2'], color=['g', 'm'], figsize=(8, 6))
plt.show()
```

## DatetimeIndex
En muchas oportunidades, las fechas son importadas hacia Pandas como textos. Estos se puede correguir convirtiendo estas fechas en indices tipo Datetime:

```python
URL = "https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20190514&end=20200514"
df = pd.read_html(URL)[2]
df
```

```python
df.index = pd.DatetimeIndex(df['Date']) #convierte string en daterange
df.drop('Date', axis=1, inplace=True)
df.head()
```

```python
#df['Close**'].plot()
df.plot(y='Close**', color='b', figsize=(8, 6)) # si no se especifícia en x siempre estará el índice
plt.show()                                       # el plot se ajusta para que muestre las fechas correctamente,
                                                    # ya que la data se convirtión en datarange
```

## Otros gráficos

Adicionalmente, se pueden obtener estos gráficos adicionales en Pandas:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist         
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

También se puede llamar a `df.plot(kind='hist')` o reemplazar el argunento `kind` con cualquier termino mostrado en la lista anterior e.g. 'box','barh', etc..)

```python
df = pd.DataFrame(data=np.random.randint(10, 100, (10, 4)))
df.rename(columns=lambda x: "Data" + str(x), inplace=True)
print(df)

df.plot.bar(y='Data0')
plt.show()
```

```python
df.plot.bar(figsize=(12, 6))
plt.show()
```

```python
df.plot.bar(stacked=True, figsize=(12, 6))  #barras apiladas
plt.show()
```

```python
df.plot.bar(subplots=True, layout=(2, 2), figsize=(12, 6))   # layout: indica la forma de la figura (2 filas,2 columnas)
plt.tight_layout()
plt.show()
```

```python
df.plot.pie(y='Data0', figsize=(12, 6), autopct="%.1f%%", ) # autopct: calculo el porcentaje de cada sección
                                                            # %.1f: especifica el formato (1 decimal)
                                                            # %%: muetra el signo de % en la gráfica (1 decimal)
plt.show()
```

```python
df[['Data0', 'Data1']].plot.pie(figsize=(12, 6), autopct="%.1f%%", subplots=True)
plt.show()
```

```python
# Histograma: conteo delas  frecuencias de valores(cuantas veces se repiten en una muestra)

df = pd.DataFrame(data=np.random.randn(1000, 3))
df.plot.hist(y=0, bins=100, color='g', figsize=(8, 6))  # y indica la columna que se quiere graficar
                                                        # bins: indica el numero de barras (en cuanto secciones se va aseparara la muestra)
plt.show()
```

```python
df = pd.DataFrame(data=np.random.randn(1000, 3))  
df.plot.hist(alpha=0.5, figsize=(8, 6))  # como no se especifica el y, entonces aparecen todas las columnas
                                        # alpha se utiliza para traslapar gráficos (indica cuán transparentes son los gráficos 0-1)
```

```python
# Gráfico de distribución o dispersión

df = pd.DataFrame(data=np.random.uniform(0, 2, (500, 3)),
                 columns=['C0', 'C1', 'C2'])
df.plot.scatter(x='C0', y='C1', figsize=(8, 6))
plt.show()
```

```python
df.plot.scatter(x='C0', y='C1', c='C2', cmap='coolwarm', figsize=(8, 6)) #C afecta el color de los puntos entre blanco y negro, mientras más lato se a c2 más blanco es el punto
plt.show()                                   # coolwarm 
                                            
```

```python
df.plot.scatter(x='C0', y='C1', s=df['C2']*200, alpha=0.5, figsize=(8, 6))   # s debe ser entero
plt.show()
```
